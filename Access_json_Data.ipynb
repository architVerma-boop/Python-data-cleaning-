```python
import json
```

```python
def load_data(filename):
    with open(filename, "r") as f:
        data = json.load(f)
        # print(data)
        return data

data = load_data("Store_data.json")
print(data)
print(type(data))
```

```python
def clean_data(data):
    text_to_num = {"one": 1, "two": 2, "three" : 3, "four": 4, "five": 5}
    cleaned_data = []
    unique_users = set()
    for user in data:
        #cleaning data
        #store in new variable raw_rating
        raw_rating = str(user["rating"]).strip().lower() #strip for trailing spaces and lower to convert into lower case
        if raw_rating in text_to_num:
            raw_rating = text_to_num[raw_rating]
        # print(raw_rating)
        user["rating"] = raw_rating

        #Inconsistency is done
        # Now Handle missing data
        raw_age = user.get("age")  #user.get("age")ðŸ‘‰ Agar "age" key exist karti hai â†’ uski value aayegiðŸ‘‰ Agar exist nahi karti â†’ None milega
        if raw_age == None:
            user["age"] = None  #Agar age missing hai, to dictionary me "age": None add ho jaata hai.
        

        #deduplication
        if(user["name"].strip() in unique_users):
            continue
        unique_users.add(user["name"])
        cleaned_data.append(user)
    return cleaned_data
        
        
```

```python
#get meaningful insights
def get_insights(data):
    tot_rating = 0
    for user in data:
        tot_rating += float(user["rating"])
    print(f"Average rating : {tot_rating/len(data)}")

    #Percentage of users with poor rating 
    poor_rating = 0
    for user in data:
        if(float(user["rating"]) < 3):
            poor_rating += 1
    print(f"Percentage of users with poor rating is : {poor_rating/len(data) *100}%")
```

```python
get_insights(data)
```
